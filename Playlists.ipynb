# Make a playlist

```python
url = f"https://api.spotify.com/v1/users/{user_id}/playlists"

headers = {"Authorization": "Bearer "+token,
          "Content-Type": "application/json"}

data = json.dumps({"name": "API-created Playlist",
                   "description": "This playlist was created using Python and Spotify API",
                   "public": "true"}, indent=4)

response = requests.post(url, headers=headers, data=data)

print(response.status_code)
print(response.json())
```

# Add songs to playlist

```python
playlist_id = "2SwAwDlKuhPXj3jqGEw8nr"

url = f"https://api.spotify.com/v1/users/{user_id}/playlists/{playlist_id}/tracks"

headers = {"Authorization": "Bearer "+token,
          "Content-Type": "application/json"}

data = json.dumps({"uris": ["spotify:track:3khEEPRyBeOUabbmOPJzAG",
                            "spotify:track:1XeOqgzh2fZhmgxnmyiIJ8"]}, indent=4)

response = requests.post(url, headers=headers, data=data)

print(response.status_code)
print(response.json())
```

# Search for item

```python
url = "https://api.spotify.com/v1/search"

headers = {"Authorization": "Bearer "+token,
          "Content-Type": "application/json"}

params = {"q": "kickback",
         "type": "track",
         "limit":1}

response = requests.get(url, params=params, headers=headers)

print(response.status_code)
print(response.json())
```
